This notebook contains an exploratory analysis of the data from jet engine

```python
import tqdm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
import os
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
```

```python
# defining a function for loading the data - add relative path!
def load_data(index="FD004"):
    if type(index) == str:
        assert index in ["FD001", "FD002", "FD003", "FD004"]
    elif type(index) == int:
        assert index in [0, 1, 2, 3]
        index = f'FD00{index+1}'

    print("-----------------")
    print(f" Data Set: {index} ")
    print("-----------------")
    if index == "FD001":
        print("Train trjectories: 100")
        print("Test trajectories: 100")
        print("Conditions: ONE (Sea Level)")
        print("Fault Modes: ONE (HPC Degradation)\n")
    if index == "FD002":
        print("Train trjectories: 260")
        print("Test trajectories: 259")
        print("Conditions: SIX")
        print("Fault Modes: ONE (HPC Degradation)\n")
    if index == "FD003":
        print("Train trjectories: 100")
        print("Test trajectories: 100")
        print("Conditions: ONE (Sea Level)")
        print("Fault Modes: TWO (HPC Degradation, Fan Degradation)\n")
    if index == "FD004":
        print("Train trjectories: 248")
        print("Test trajectories: 249")
        print("Conditions: SIX")
        print("Fault Modes: TWO (HPC Degradation, Fan Degradation)\n")

    train_set = np.loadtxt(f"train_{index}.txt")
    test_set  = np.loadtxt(f"test_{index}.txt")

    col_names = ["unit_number", "time"]
    col_names += [f"operation{i}" for i in range(1, 4)]
    col_names += [f"sensor{i}" for i in range(1, 22)]
    train_set = pd.DataFrame(train_set, columns=col_names)
    test_set  = pd.DataFrame(test_set, columns=col_names)
    labels = np.loadtxt(f"RUL_{index}.txt")

    def set_dtype(df):
        return df.astype({"unit_number": np.int64, "time": np.int64})

    train_set = set_dtype(train_set)
    test_set  = set_dtype(test_set)

    return train_set, test_set, labels

```

```python
# reading in the data using previously defined function
train_set, test_set, labels = load_data(index=3)
```

```python
# investigate the input data
#%whos - command to check available variables
train_set.head()
```

In order to investigate the changes in the units, it is important to compare the operation at the same or similar conditions to rule out the influence of the conditions on the results. The operating settings are given with three variables, by plotting those variables it is possible to see the extent and number of the conditions. By plotting the conditions in respect to time, it will be possible to see it they are discrete or some transitions from one to another are also represented in the data. 

```python
fig, axs = plt.subplots(1,3)
axs[0].scatter(train_set['time'], train_set['operation1'])
axs[1].scatter(train_set['time'], train_set['operation2'])
axs[2].scatter(train_set['time'], train_set['operation3'])
#plt.scatter(train_set['time'], train_set['operation2'])
```

By simple examination, it is possible to see that there is a finite number of operational settings provided and they represent discrete conditions. Those three variables can form combinations of conditions. In order to find how many combinations have to be considered, the scatter plot of the conditions is plotted.  

```python
# plotting operating conditions for different regimes
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(train_set['operation1'],train_set['operation2'],train_set['operation3'])

ax.set_xlabel('Op 1')
ax.set_ylabel('Op 2')
ax.set_zlabel('OP 3')
```

By investigating the conditions, 6 distinct operating regimes can be found in the data. The variability can be investigated within each of the regimes, but for the time being it is assumed that points from each cluster represent the same conditions. 

```python
# inestigation of the discrepancy within clusters - left out for now
np.shape(train_set['operation1'].unique())
df = train_set['operation1'].unique()
df = df[df<5]
counts, bins = np.histogram(df)
plt.stairs(counts, bins)

np.shape(train_set['operation2'].unique())
df = train_set['operation2'].unique()
df = df[df<5]
counts, bins = np.histogram(df)
plt.stairs(counts, bins)

plt.plot(df)
```

```python
X1 = np.array(train_set['operation1'])
X2 = np.array(train_set['operation2'])
X3 = np.array(train_set['operation3'])
X = np.transpose(np.stack([X1,X2,X3]))
np.shape(X)
```

```python
# predicting clusters based on the input data
y_pred = KMeans(n_clusters=6).fit_predict(X)
```

```python
# add column with cluster number and see if it was added correctly
train_set["class"] = y_pred
train_set.head()
```

```python
# investigating if any of the regimes was the most common
df_mean = train_set.groupby(["class"]).mean()
# the time for condition 4 is the highest, other than that it is almost identical
df_std = train_set.groupby(["class"]).std()
train_set.groupby(["class"]).sum()
```

Is it better to use different conditions or combine the conditions all together? Is the delection of conditions related to the accelerated degradation of the engine?

```python
# check if each engine has the same conditions most of the time
operation_per_class = []
sum_of_cycles = []
for i in range(100):
    df_temp = train_set[train_set["unit_number"] == i+1]
    operation_per_class.append(df_temp.groupby(["class"]).sum()["unit_number"]/(i+1))
    sum_of_cycles.append(sum(operation_per_class[i]))
    operation_per_class[i] = operation_per_class[i]/sum(operation_per_class[i])
```

```python
sum_of_cycles[1]
```

```python
plt.plot(operation_per_class)
```

```python
np_operation_per_class = pd.concat(operation_per_class,axis=1).values
#np_operation_per_class = np.reshape(np_operation_per_class,(100,6))
np_operation_per_class.shape
```

```python
correlations = []
for i in range(6):
    correlations.append(np.corrcoef(sum_of_cycles,np_operation_per_class[i,:])[1,0])
correlations
```

```python
#distribution of conditions
for i in range(6):
    plt.hist(x=np_operation_per_class[i,:], bins='auto', color='#0504aa', alpha=0.7, rwidth=0.85)
```
